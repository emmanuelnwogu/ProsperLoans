## Insight 6: Is there a relationship between borrowers' APR and interest rates?

```python
sns.regplot(data=df,x='BorrowerAPR',y='BorrowerRate',color='blue',fit_reg=False); #fit_reg=False removes the regression line
plt.xlabel('Borrower APR',color='brown');
plt.ylabel('Borrower Interest Rate',color='brown');
plt.title('Correlation between Borrower APR and Interest Rates',color='brown');
```

### --The above scatter plot demonstrates a strong positive correlation between the APR anbd the interest rate for prosper's borrowers

## Conclusions/Summary of Findings
From the above data wrangling and analysis, we can reach the following conclusions:

- Over the years,overall, Prosper firm granted more and more loans to borrowers.

- Borrowers with a high debt-to-income ratio are more likely to default/delay in repayment. This means we should target more people with a lower ratio.

- Prosper firm gives a lower APR to borrowers who already have a home. The reason for this is most likely because they are more likelyu to remain committed to the contract. 

- Borrowers with credit grades AA,A and B accrue the largest volume of loans as they require the highest number of investors to fund their loans.
